```python
# Code No.1
import pandas as pd
brainFile = '/content/brainsize.txt'
brainFrame = pd.read_csv(brainFile, '\t')
```

```python
# Code No.2
brainFrame.head()
```

```python
# Code No.3
brainFrame.describe()
```

```python
# Code No.4
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Code No.5
menDf = brainFrame[(brainFrame.Gender == 'Male')]
womenDf = brainFrame[(brainFrame.Gender == 'Female')]
```

```python
# Code No.6
menMeanSmarts = menDf[["PIQ", "FSIQ", "VIQ"]].mean(axis=1)
plt.scatter(menMeanSmarts, menDf["MRI_Count"])
plt.show()
%matplotlib inline
```

```python
# Code No.7
womenMeanSmarts = womenDf[["PIQ", "FSIQ", "VIQ"]].mean(axis=1)
plt.scatter(womenMeanSmarts, womenDf["MRI_Count"])
plt.show()
%matplotlib inline
```

```python
# Code No.8
brainFrame.corr(method='pearson')
```

Notice at the left-to-right diagonal in the correlation table generated above. Why is the diagonal filledwith 1s? Is that a coincidence? Explain.

The reason why the correlation table are filled with diagonal 1s is due to Pearsons Correlation Coefficient as it measures a linear correlation between two continuous variables in a -1 and 1 measurement.
