```python
#Code No.1

import pandas as pd

training = pd.read_csv("/content/titanic_train.csv")
```

```python
#Code No.2

training.info()
```

```python
#Code No.3

training.head()
```

```python
#Code No.4

training["Gender"] = training["Sex"].apply(lambda toLabel: 0 if toLabel == 'male' else 1)
```

```python
#Code No.5

training.head()
```

```python
#Code No.6

training["Age"].fillna(training["Age"].mean(), inplace=True)
```

```python
#Code No.7

training.info()
```

```python
#Value that is replaced to the missing ages

print(training["Age"].mean())
```

```python
#Code No.8

y_target = training["Survived"].values
```

```python
#Code No.9

columns = ["Fare", "Pclass", "Gender", "Age", "SibSp"]
X_input = training[list(columns)].values
```

```python
#Code No.10

from sklearn import tree

clf_train = tree.DecisionTreeClassifier(criterion="entropy", max_depth=3)

clf_train = clf_train.fit(X_input, y_target)
```

```python
#Code No.11

score = clf_train.score(X_input,y_target)
print(round(score*100, 2),"%")
```

```python
#Code No.12

from six import StringIO
with open("/content/titanic_train.csv", 'w') as f:
  f = tree.export_graphviz(clf_train, out_file=f, feature_names=columns)
```
